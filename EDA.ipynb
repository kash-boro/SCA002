EXPLORATORY DATA ANALYSIS:

EDA is the process of understanding data using statisticsand visualizations before building machine learning models. 

steps of EDA:

Collect Data
↓
Understand Dataset
↓
Check Missing Values
↓
Handle Missing Values
↓
Remove Duplicates
↓
Detect Outliers
↓
Data Visualization
↓
Feature Relationships
↓
Statistical Summary
↓
Ready for ML
